### Matplotlib Interfaces

Matplotlib offers two types of plotting interfaces. The first interface is the `matplotlib.pyplot` interface or the MATLAB-style state-based interface. The second interface is the `matplotlib.pyplot.figure` interface or the object-oriented interface. It is a more powerful interface that allows you to create figures and plot on them directly.

Let's explore each of them in detail.

#### `Matplotlib.pyplot` or MATLAB style Interface

Matplotlib was originally written as a python alternative to MATLAB. The MATLAB-style interface is still supported in matplotlib. The `matplotlib.pyplot` interface is a state-based interface that allows you to create figures and plot on them directly.

Let's look at an example of how to use the `matplotlib.pyplot` interface.

````python
# First create a plot figure
plt.figure()

# Create two subplots and plot the sine and cosine functions
# Plot the sine function
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
plt.plot(x, np.sin(x))

# Plot the cosine function
plt.subplot(2, 1, 2) # 2 rows, 1 column, second subplot
plt.plot(x, np.cos(x))
````

In the above code, we have first initialised a figure object, and then created subplots using the `subplot` method. The `subplot` method takes three arguments, `nrows`, `ncols` and `index`. The `nrows` argument is the number of rows in the subplot grid, and the `ncols` argument is the number of columns in the subplot grid. The `index` argument is the index of the subplot in the grid.After creating the subplots, we have plotted the `sine` and `cosine` functions on the subplots.

It is important to know that this interface is `stateful` i.e. it keeps track of the current figure and axes, which are all the `plt` plots are applied. You cen get the current figure and axes by calling the `plt.gcf()` and `plt.gca()` methods.

This state interface is fast and convinient, but it is not very flexible. Once you have created the second subplot, you cannot easily go and edit the first subplot. This interface is easy and convienient for smaller plots, but for large plots, we can use the `matplotlib.pyplot.figure` interface.

#### `Matplotlib.pyplot.figure` or Object-oriented Interface

The object-oriented interface is available for more complicated situations, when we want more control over the figure and axes. Rather than depending on some notion of an "active" figure or axes, in the object-oriented interface the plotting functions are methods of explicit Figure and Axes objects. To re-create the previous plot using this style of plotting, you might do the following:

````python
# First create the grid of subplots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call the plot() method on the appropriate Axes object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
````

As we can see, the plotting became more easy to control i.e. now we can easily edit the subplots by indexing the axes.

Cool! Now since we have got some basic notion of plotting, let's move to the next section on `Basic Plotting with Matplotlib`.

## Visualization using Matplotlib

In this section, let's explore various types of plots that can be created using the `matplotlib.pyplot` interface. Let's now get started with the `line` plot.

### Line Plot

A line plot is a plot in which the points are connected by a line. The `plot` method of the `matplotlib.pyplot` interface takes two arguments, `x` and `y`. The `x` argument is the x-axis values, and the `y` argument is the y-axis values.It is the simplest of the plotting methods, that provides a visualisation of a single function `y=f(x)`.

Before we can plot a line, we need to create a figure and axes object.

````python
# Set the style of white-grid
plt.style.use('seaborn-whitegrid')

# Create a figure and axes
fig = plt.figure()
ax = plt.axes()
````

In Matplotlib, the `figure` object consists of all the objects representing axes, graphics, text and labels. The `axes` object is the object representing the axes of the figure i.e. the bounding box with ticks and labels, which is the plotting area. Once we have created the figure and axes object, we can plot in the axes using the `ax.plot()` method.

Now let's plot a line using the same.

````python
# Create the figure and axes
fig = plt.figure()
ax = plt.axes()

# create the x-axis array of 100 values between 0 to 10
x = np.linspace(0, 10, 100)
# Now plot the sine function
ax.plot(x, np.sin(x))
````

Alternatively, we can use the `plt.plot()` method so that the figure and the axes are created automatically.

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
````

Also note that we can use the `plt.plot()` method multiple times to plot multiple lines on the same figure. Now let's use some methods to customize the plot.

#### 1. Adjusting Colors of the plot

We can change the color of the plot by using the `color` argument. The `color` argument is a string that can be any of the colors supported by matplotlib. The default color is `'b'`. Other ways to specify the color are:
* Specifying the color using the hexadecimal code.
* Specifying the color using the name of the color.
* Specifying the color using the RGB values.

Let's check the supported colors by calling the `get_named_colors_mapping` method on the `matplotlib.colors` module.

```python
import matplotlib.colors as colors
print(colors.get_named_colors_mapping())
```

Let's see an example,

````python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
````

#### 2. Adjusting the line style

We can change the line style by using the `linestyle` argument. The `linestyle` argument is a string that can be any of the line styles supported by matplotlib. The default line style is `'-'`. Other ways to specify the line style are:
* Specifying the line style using the name of the line style.
* Specifying the line style using the dash tuple.

Let's see an example,


````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

We can combine both the `color` and the `linestyle` by using a non-keyword argument to the `plot` method.

````python
plt.plot(x, x + 0, '-g')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k')
plt.plot(x, x + 3, ':r')
````

#### Adjusting the limits of the plot

We can adjust the limits of the plot by using the `xlim` and `ylim` arguments. The `xlim` argument is a tuple of two values, the lower and the upper limit of the x-axis. Similarly, the `ylim` argument is a tuple of two values, the lower and the upper limit of the y-axis. Let's see an example,

````python
# plot without limits
plt.plot(x, np.sin(x))
````

````python
# plot with x-limits and y-limits
plt.plot(x, np.sin(x), color='black')
plt.xlim(0, 10)
plt.ylim(-1, 1.01)
````

We can see that the plot is now perfectly fit in the figure.

#### 4. Adding title and legends

We can specify the title and the legend to the plot by using the `plt.title()` and `plt.legend()` methods. The `title` argument is a string that is the title of the plot. The `legend` argument is a list of strings that are the legends of the plot.

````python
# Add title and legend to plot
plt.plot(x, np.sin(x), color='black')
plt.xlim(0, 10)
plt.ylim(-1, 1.01)
plt.title('Sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend(['sin(x)'])
````

### Scatter Plot

Scatter Plot is similar to line plot, instead of being joined by line segments it is joined by points. The `scatter` method of the `matplotlib.pyplot` interface takes two arguments, `x` and `y`. The `x` argument is the x-axis values, and the `y` argument is the y-axis values.

Let's plot a scatter plot.

#### Using Line Plot

We can plot scatter plot using the 'plot' method of the `matplotlib.pyplot` interface and specifying the point by a dot or circle. Let's plot a scatter plot using the 'plot' method.

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
````

````python
plt.plot(x, y, '-o', color='black')
````

In the above two plots, we can observe that the scatter plot is not very different from the line plot. The only difference is that the points in the scatter plot are not connected by a line. Now let's check the `plt.scatter()` method.

#### Using the `plt.scatter()` method

We can directly use the `plt.scatter()` method to plot a scatter plot. The `plt.scatter()` method takes three arguments, `x`, `y` and `s`. The `x` argument is the x-axis values, and the `y` argument is the y-axis values. The `s` argument is the size of the points.

````python
# Scatter plot with color
plt.scatter(x, y, color='black')
````


The primary differenc of `plt.scatter()` and `plt.plot()` is that the `plt.scatter()` can be used to create scatter plots where the properties of the individual points can be easily controlled or modified according to data.

To illustrate the advantage of the `plt.scatter()` method, let's plot a scatter plot using the `plt.scatter()` method on data points of different sizes and colors. In the below example, we are plotting 100 points with different sizes and colors. By specifying the `alpha` argument, we can control the transparency of the points.

````python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
````

As we can see from the above plot, the scatter plot can specify the properties i.e. color, size etc of the individual points. This is very useful when we want to plot the dataset that has different properties for each data point.

#### `plot()` vs `scatter()`: Which one to use ?

Now we have seen the chracteristics of both the plotting methods. But if we have to apply the scatter plot we need to remeber the following:

* Use the `plt.scatter()` method when the size of dataset is small. Because the `plt.scatter()` method has the capabilites to control the properties of the individual points. So if the dataset is large, we should use the `plt.plot()` method.

* Use the `plt.plot()` method when the size of dataset is large. Because the `plt.plot()` method has the capabilites to control the properties of the entire plot. It will consider every point to be of equal size and it will be easier and faster to use the `plt.plot()` method on large datasets.

### Bar Plot

A Bar plot is a graph that represents data with rectangular bars with heights proportional to the values that they represent. The `bar` method of the `matplotlib.pyplot` interface takes two arguments, `x` and `y`. The `x` argument is the x-axis values, and the `y` argument is the y-axis values. It also accepts parameters like `color`, `alpha`, `width` etc that can be used to control the properties of the bars. Let's look at an example.

````python
# Create data for bar plot
# Set seed and get data
np.random.seed(0)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# Plot the bar plot
fig =  plt.figure()
ax = plt.axes()

ax.bar(x, y, width=1, color='black', edgecolor='white', linewidth=2)
ax.set(xlim=(0, 8), xticks=np.arange(0, 8),
       ylim=(0, 8), yticks=np.arange(0, 8));
plt.show()
````

In the above graph, we can see that the bars are of different sizes depending on the data. By specifying the `width` argument, we can control the width of the bars. Some of the other parameters that can be used are `color`, `alpha`, `align`, `edgecolor`, `linewidth` etc.

The `Bar` plot is best used when we have to visualize discrete data. For example, if we have to visualize the number of students in different classes. This will be best useful to check the class distribution in a classification dataset.

### Box Plot

A Box plot is a graphical method for displaying numerical data with five-number summaries:

* The minimum: the smallest data value.
* First quartile (Q1): the 25th percentile (the point at which 25% of the data are below).
* Median (Q2): the 50th percentile (the point at which 50% of the data are below).
* Third quartile (Q3): the 75th percentile (the point at which 75% of the data are below).
* The maximum: the largest data value.

The boxplot is used to demostrate the locality, spread and skewness of numerical data through their quartiles.

The `boxplot` method of the `matplotlib.pyplot` interface takes one argument, `x`. The `x` argument is the x-axis values, and the `y` argument is the y-axis values. It also accepts parameters like `color`, `alpha`, `width` etc that can be used to control the properties of the bars. Let's look at an example.

````python
# Example of box plot

# Set seed
np.random.seed(0)

# Get 3 random arrays from the normal distribution
x1 = np.random.normal(100, 10, 100)
x2 = np.random.normal(90, 20, 100)
x3 = np.random.normal(80, 30, 100)
x = [x1, x2, x3]

# Plot the boxplot
fig = plt.figure()
ax = plt.axes()

bp = ax.boxplot(x, )
plt.show()
````

In the above code, we have created three array of size 100 from normal distribution but with different means and standard deviations. Then created the boxplot using these arrays. We can see that the boxplot is very useful when we have to visualize the statistics of the data.
For the above boxplot, we can find the following observations:

* The `median` of the data is represented by the orange line in the center of the boxplot.
* The `first quartile (Q1)` is represented by the line below the median.
* The `third quartile (Q3)` is represented by the line above the median.
* The `minimum` and `maximum` of the data is represented by the lines at the extreme ends of the boxplot.
* The `whiskers` are the lines extending from the boxplot to the data.
* The `fliers` or `outliers` are the points that are not in the boxplot.

Thus if we observe the boxplot, we will be able to easily understand the distribution of the data.

## Animations using Matplotlib

Animations are an interesting and powerful way to visualize data. Compared to static charts, animations are much more interactive and can be used to create dynamic visualizations.

In this section, we will explore how to create animations using the `matplotlib.animation` module. It provides a framework around which the animation can be built. It provides two interfaces to create the animation: `FuncAnimation` and `ArtistAnimation`. In this section, we will look at the `FuncAnimation` interface to animate the sine wave.

#### Basic Animation - Sine Wave

Let's use the `Funcanimation` interface to animate the sine wave. Plotting animation involves the following steps:
* Create the figure.
* Create the axes.
* Create the line object.
* Define the `init()` function.
* Define the `animate()` function.

Now let's do the same in code.

````python
from matplotlib.animation import FuncAnimation, PillowWriter
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-pastel')

# 1. Create figure and axes
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))

# Create the line object
line, = ax.plot([], [], lw=2)

# Define the init() function
def init():
    line.set_data([], [])
    return line,

# Define the animation function
def animate(i):
    x = np.linspace(0, 4*i/200, 1000)
    y = np.sin(2 * np.pi * x)
    line.set_data(x, y)
    print(i)
    return line,

# Define the animation object
sine_animation = FuncAnimation(fig, animate, init_func=init, frames=200, interval=20, blit=True)

# Start animation
sine_animation.save('sine_animation.gif', writer=PillowWriter(fps=30))
````

In tha above code, we have followed the following steps:

* Create the figure: Create the figure object using the `plt.figure()` function.
* Create the axes: Create the axes object using the `plt.axes()` function.
* Create the line object: Create the line object using the `plt.plot()` function. This object will be updaated evertime with new data and will be displayed on the axes.
* Define the `init()` function: This function is called once to set up the plot. In the above code, we have defined the line object with empty x and y values.
* Define the `animate()` function: This function is called everytime the figure is redrawn and is responsible for updating the line object. In the above code, we have defined the line object with new x and y values i.e. the sine wave. The `blit=True` parameter is used to update the line object only when the data is changed.
* Call the `animation.FuncAnimation()` function: Here we will initialise the `FuncAnimation` with the `init()` function and the `animate()` function. We will also pass the figure, axes and line object to the `FuncAnimation()` function.
* Save the animation: We can save the animation using the `animation.save()` function. We should provide the name of the file to save the animation and the writer to use.

Great! Now we have created an animation of sine wave and saved it as a gif file. Open and check the `GIF` file to see the animation.
